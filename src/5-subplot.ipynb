### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = x ** 2  # Quadratic values for line plot and scatter plot
bar_values = [5, 7, 3, 8, 4]
labels = ['A', 'B', 'C', 'D', 'E']
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create subplots (2x2)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot (1st subplot)
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Line Chart")

# Bar Chart (2nd subplot)
axes[0, 1].bar(labels, bar_values, color='g')
axes[0, 1].set_title("Bar Chart")

# Scatter Plot (3rd subplot)
axes[1, 0].scatter(x, y, color='r', marker='o')
axes[1, 0].set_title("Scatter Plot")

# Pie Chart (4th subplot)
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
